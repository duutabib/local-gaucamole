```python
def orlicz(x, psi, ts = logspace(-3, 3, 10000)):
    for t in ts:
        d = psi(abs(x)/t)
        if mean(d) <= 2:
            return t      
    return np.inf

def subG(x, **qargs):
    f = lambda x: exp(x**2) -1 
    return orlicz(x, f, **qargs)

def subE(x, **qargs):
    f = lambda x: exp(x) -1
    return orlicz(x, f, **qargs)
```
