```python
def f(x):
    return x**4*np.exp(-x)
```

```python
x =  np.linspace(-3, 3, 101)
y = f(x)
```

```python
def f1(t):
    return t**2*exp(-t**2)

def f2(t):
    return t**2*f1(t)

t = linspace(0, 3, 51)
y = f1(t)
y0 =f2(t)

plot(t, y)
plot(t, y0)
xlabel("t Values")
ylabel(f"t^2*exp(-t^2)")
legend([f"t^2*exp(-t^2)", "t^4*exp(-t^2)"])
axis([0, 3, -0.05, 0.6])  # [tmax, tmin, ymin, ymax]
title(" Plotting two curves f(t) for the [0, 3]")
show()
```

# Plots in the same figure

```python
figure()
subplot(2, 1, 1)
t = linspace(0, 3, 51)
y1 = f1(t)
y2 = f2(t)

plot(t, y1, 'r-', t, y2, 'bo')
xlabel('t')
ylabel('y')
axis([t[0], t[-1], min(y2)- 0.05, max(y2) + 0.5])
legend([f"t^2*exp(-t^2)", f"t^4*exp(-t^2)"])
title("Top Figure")


subplot(2, 1, 2)
t3 = t[::4]
y3 =f2(t3)

plot(t, y1, 'b-', t3, y3, 'ys')
xlabel('t')
ylabel('y')
axis([0, 4, -0.2, 0.6])
legend([f"t^2*exp(-t^2)", f"t^4*exp(-t^2)"])
title("")
```
