```python
%pylab inline
from scipy.sparse.linalg import eigsh

rcParams['figure.figsize'] = (15, 5)
```

```python
def rend_cov(x, *kwargs):
    xe, xv = eigsh(x, k=2)
    #xe_ = np.sqrt(abs(sort(xe)))
    #xe_ = xe_[:2, ]
    t = linspace(0, 2*pi, 100)
    t_ = np.array([cos(t), sin(t)])
    es = xv @ (xe.reshape(2, 1) * t_) 
    plot(es[0], es[1], color ='white' )
    

    return None
```

# Plotting ellipse on gaussian distributions

```python
xx = randn(2, 1000000)
A = array([[2, 1], [1, 1.5]])
yy = A @ xx
tp = cov(yy)
print(tp.shape)
```

```python
hist2d(yy[0], yy[1], bins=1000)
#axis('equal')
rend_cov(tp)
# suppress the hist2d
```
